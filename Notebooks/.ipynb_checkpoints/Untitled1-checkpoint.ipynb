```python
ncluster_list = range(1,10)

def kmean_score(nclust):
    km = KMeans(n_clusters = nclust, 
                init = 'random', 
                n_init = 10, 
                max_iter = 300, 
                n_jobs = -1) #look at hyperparameters
    km.fit(X)
    rss = -km.score(X)
    return rss

scores = [kmean_score(i) for i in range(1,10)]
print("#clust\trss")
for i, score in enumerate(scores):
    print("{0}\t{1:0.0f}".format(ncluster_list[i],score))
```

```python
def ho_testing():
    cols=['Second','Third','9999.0', 'Camp Length']
    check = []
    for i in cols:
        
        ho = dff[(dff[i]) & (dff['y_target']==1) ]
        ha = dff[(dff[i]) & (dff['y_target']==0) ]
        print(sum(ha[i].values) , len(ha), sum(ho[i].values) , len(ho))
        ho_mean = ho.mean()
        ha_mean = ha.mean()
        
        results = sm.stats.proportions_ztest(ho_mean,ha_mean , alternative = 'two-sided')
        check.append((i,results))
    return check

print(ho_testing())
```
